```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_name = 'camera.png'
```

```python
image = cv2.imread(image_name)

plt.imshow(image)
```

```python
def apply_perspective_transform(src_points, dst_points, input_image, dest_image_size):
    # Convert points to numpy arrays
    src_points = np.array(src_points, dtype=np.float32)
    dst_points = np.array(dst_points, dtype=np.float32)

    # Calculate the perspective transformation matrix
    matrix = cv2.getPerspectiveTransform(src_points, dst_points)

    # Apply the perspective transformation
    transformed_image = cv2.warpPerspective(input_image, matrix, dest_image_size)

    return transformed_image
```

```python
width, height = 1200, 900
dst_points = np.float32([[0, 0], [width, 0], [0, height], [width, height]])
src_points = np.float32([[150, 175], [440, 111], [223, 395], [528, 293]])
```

```python
image_noperspective = apply_perspective_transform(src_points, dst_points, image, (width, height))
plt.imshow(image_noperspective)
```

```python
cv2.imwrite(image_name + '_.jpg', image_noperspective)
```
